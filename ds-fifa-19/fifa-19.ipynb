# Data Science Study - FIFA 19
Player Recommender System based on similar player

Dataset from : https://www.kaggle.com/karangadiya/fifa19

```python
from IPython.core.display import display, HTML
import warnings
warnings.filterwarnings('ignore')
```

### 1- Data Extraction

```python
import pandas as pd

df = pd.read_csv("./fifa-19-dataset.csv")
df.head()
```

### 2- Data Preparation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler


fifa19DF = df[['ID','Name','Nationality','Club','Position','Overall','Potential','Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']]

fifa19DF = fifa19DF.dropna()

positionEncoder = LabelEncoder().fit(fifa19DF['Position'])
fifa19DF['Position'] = positionEncoder.transform(fifa19DF['Position'])

fifa19KNN = fifa19DF[['Position','Overall','Potential','Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']]
```

### 3- Model Creation & Evaluation

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(11,metric='manhattan').fit(fifa19KNN)

playerID = 190871
player = fifa19KNN[fifa19DF['ID']==playerID]
top_k_distances, top_k_players = model.kneighbors(player)

playerData = fifa19DF[fifa19DF['ID']==playerID]
playerResult = fifa19DF[fifa19DF.index.isin(top_k_players[0][1:])]
playerData['Position'] = positionEncoder.inverse_transform(playerData['Position'])
playerResult['Position'] = positionEncoder.inverse_transform(playerResult['Position'])
```

```python
display(HTML("<h3>Player: </h3>"+playerData.to_html()))
```

```python
display(HTML("<h3>Players Like "+ playerData['Name'].item() + ":</h3>"+ playerResult.to_html()))
```
